```python
import os
from glob import glob
import cv2
import matplotlib.pyplot as plt
import csv
import random
import pandas as pd
```

```python
path = '/data2/hang/SIA/SIA_luggage_damages/MissingWheel/context_learning_data/luggage_parts_annotations_total_2609_newbody'
```

```python
files = glob(os.path.join(path,'*','*'))
print(len(files))
```

```python
positive_files = [f for f in files if 'pos_' in f and '_raw' in f]
pos_labels = [1 for f in positive_files]
print(len(positive_files))
```

```python
negative_files = [f for f in files if 'neg_' in f]
neg_labels = [0 for f in negative_files]
print(len(negative_files))
```

```python
neg neg -> model
raw mask -> model 
```

```python
df = {'file_name':positive_files+negative_files,
     'label':pos_labels+neg_labels}
df = pd.DataFrame(df)
df.to_csv('total_data.csv',index=False)
```
